```python
df["Sampling Timestamp"] = pd.to_datetime(
    df["Sampling Timestamp"], errors="coerce")

df["Sampling Date"] = df["Sampling Timestamp"].dt.date
df.drop(columns = ["Sampling Timestamp"], inplace = True)

```

```python
df
```

```python
exclude_cols = ["Distance Group", "Sampling Timestamp", "Site Name", "Unit Symbol"]

unit_map = {}

for col in df1.columns:
    if col in exclude_cols:
        continue
    
    # find rows where this parameter has a value
    mask = df1[col].notna()
    
    if mask.sum() > 0:
        # get most common unit
        unit = df1.loc[mask, "Unit Symbol"].mode()[0]
        unit_map[col] = unit

# print nicely
for param, unit in unit_map.items():
    print(f"{param}: {unit}")
```

```python
units = {
    "Aluminum Dissolved": "µg/L",
    "Aluminum Total Recoverable": "µg/L",
    "Calcium Dissolved": "mg/L",
    "Calcium Total": "mg/L",
    "Copper Dissolved": "µg/L",
    "Copper Total Recoverable": "µg/L",
    "Lead Dissolved": "µg/L",
    "Lead Total Recoverable": "µg/L",
    "Mercury Dissolved": "µg/L",
    "Mercury Total": "ng/L",
    "Methyl Mercury": "ng/L",
    "Nickel Dissolved": "µg/L",
    "Nickel Total Recoverable": "µg/L",
    "Nitrogen Kjeldahl Dissolved": "mg/L",
    "Nitrogen Kjeldahl Total": "mg/L",
    "pH": "pH units",
    "pH (Field)": "pH units",
    "Phosphorus Total": "mg/L",
    "Phosphorus Total Dissolved": "mg/L",
    "Turbidity": "NTU",
    "Vanadium Total Recoverable": "µg/L",
    "Vanadium Dissolved": "µg/L",
    "OXYGEN BIOCHEMICAL DEMAND": "mg/L",
    "Oxygen Dissolved (Field Meter)": "mg/L",
    "Oxygen dissolved % saturation": "%"
}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

sns.set(style="whitegrid")

os.makedirs("boxplots", exist_ok=True)

exclude_cols = ["Year", "Month"]

variables = [
    col for col in df.select_dtypes(include=np.number).columns
    if col not in exclude_cols
]

for var in variables:
    plt.figure(figsize=(6,4))
    
    sns.boxplot(
        data=df,
        x='Distance Group',
        y=var
    )

    sns.stripplot(
        data=df,
        x='Distance Group',
        y=var,
        color='black',
        alpha=0.2,
        size=2
    )

    if unit:
        plt.title(f'{var} by Distance Group ({unit})')
        plt.xlabel('')  
    else:
        plt.title(f'{var} by Distance Group')
        plt.xlabel('')
    plt.ylabel(var)

    plt.tight_layout()

    boxname = f"boxplots{var.replace(' ', '_')}.png"
    plt.savefig(boxname, dpi=300)

    plt.close() 
```

```python
exclude_cols = ["Year"]  # keep Distance Group for grouping
variables = [
    col for col in df.select_dtypes(include=np.number).columns
    if col not in exclude_cols
]


for var in variables:
    unit = units.get(var, "")
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=var, hue='Distance Group', bins=20, kde=True)
    if unit:
        plt.title(f'Distribution of {var} ({unit})')
        plt.xlabel(f'{var} ({unit})')  
    else:
        plt.title(f'Distribution of {var}')
        plt.xlabel(var)

    plt.ylabel('Frequency')    
    plt.tight_layout()
    barname = f"barplots{var.replace(' ', '_')}.png"
    plt.savefig(barname)
    plt.show()
```

```python
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'])
df['Year'] = df['Sampling Date'].dt.year
```

```python
exclude_cols = ["Year"]  # keep Distance Group for grouping
variables = [
    col for col in df.select_dtypes(include=np.number).columns
    if col not in exclude_cols
]

for var in variables:
    plt.figure(figsize=(6,4))
    sns.lineplot(data=df, x='Year', y=var,
                 hue='Distance Group', estimator='mean')
    if unit:
        plt.title(f'Lineplot of {var} ({unit})')
        plt.xlabel(f'date')  
    else:
        plt.title(f'Lineplot of {var}')
        plt.xlabel('date')    
        plt.tight_layout()
    linename = f"lineplots{var.replace(' ', '_')}.png"
    plt.savefig(linename)
    plt.show()
```
