```python
import numpy as np
import pandas as pd

namur = pd.read_csv("NamurLake_Data.csv")
Touchwood = pd.read_csv("Touchwood_Data.csv")
```

```python
Touchwood
```

```python
df = pd.concat([namur, Touchwood])
```

```python
df = df.drop(columns=["Unnamed: 0"], errors = "ignore")
df["Sampling Timestamp"] = pd.to_datetime(
    df["Sampling Timestamp"], errors="coerce"
)
df = df.sort_values(by="Sampling Timestamp")
df["Distance Group"] = "Far-Fields"
df
```

```python
import matplotlib.pyplot as plt
df.hist(figsize=(20,16))
plt.show()
```

```python
import matplotlib.pyplot as plt

chem_cols = [
    "Aluminum Dissolved",
    "Aluminum Total Recoverable",
    "Calcium Dissolved",
    "Calcium Total",
    "Copper Dissolved",
    "Copper Total Recoverable",
    "Lead Dissolved",
    "Lead Total Recoverable",
    "Nickel Dissolved",
    "Nickel Total Recoverable",
    "Nitrogen Dissolved",
    "Phosphorus Total Dissolved",
    "Phosphorus Total",
    "Vanadium Dissolved",
    "Turbidity",
    "pH",
    "Oxygen dissolved % saturation",
    "OXYGEN BIOCHEMICAL DEMAND"
]

fig, axes = plt.subplots(5, 4, figsize=(25,15))
axes = axes.flatten()

for i, col in enumerate(chem_cols):
    axes[i].boxplot(df[col].dropna())
    axes[i].set_title(col)
    axes[i].set_ylabel("Concentration")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.dates as mdates

fig, axes = plt.subplots(5, 4, figsize=(25,15))
axes = axes.flatten()

for i, col in enumerate(chem_cols):
    ax = axes[i]
    
    for lake in df["Site Name"].unique():
        lake_data = df[df["Site Name"] == lake]
        
        ax.scatter(
            lake_data["Sampling Timestamp"],
            lake_data[col],
            label=lake
            
        )
       

    ax.xaxis.set_major_locator(mdates.AutoDateLocator(minticks=5))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    ax.tick_params(axis='x', rotation=45)
    
    ax.set_title(col)
    ax.set_xlabel("Date")
    ax.set_ylabel("Concentration")

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc="upper right")

plt.tight_layout()
plt.show()
```
